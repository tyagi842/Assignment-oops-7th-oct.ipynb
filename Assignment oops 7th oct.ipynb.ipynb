```python
# Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.
```

```python
class BankAccount:
    def __init__(self, account_number, account_holder_name, balance=0):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            return f"Deposited {amount} into the account. New balance is {self.balance}"
        else:
            return "Please enter a valid deposit amount."

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            return f"Withdrew {amount} from the account. New balance is {self.balance}"
        elif amount > self.balance:
            return "Insufficient funds. Withdrawal not allowed."
        else:
            return "Please enter a valid withdrawal amount."

    def get_balance(self):
        return f"Account balance for {self.account_holder_name}: {self.balance}"

    def __str__(self):
        return f"Account Number: {self.account_number}\nAccount Holder Name: {self.account_holder_name}\nBalance: {self.balance}"

if __name__ == "__main__":
    account1 = BankAccount("123456789", "Harsh Tyagi", 1000)

    print(account1)  # Display account information

    print(account1.deposit(500))  # Deposit 500
    print(account1.get_balance())  # Display account balance

    print(account1.withdraw(300))  # Withdraw 300
    print(account1.get_balance())  # Display account balance

    print(account1.withdraw(1500))  # Attempt to withdraw more than the balance

    print(account1.deposit(-100))  # Attempt to deposit a negative amount
```

```python
# Problem 2: Employee Management Create a class representing an employee with attributes like employee ID, name, and salary. Implement methods to calculate the yearly bonus and display employee details.
```

```python
class Employee:
    def __init__(self, employee_id, name, salary):
        self.employee_id = employee_id
        self.name = name
        self.salary = salary

    def calculate_yearly_bonus(self, bonus_percentage):
        if bonus_percentage >= 0:
            bonus = (bonus_percentage / 100) * self.salary
            return f"Yearly bonus for {self.name} (Employee ID: {self.employee_id}): {bonus}"
        else:
            return "Please enter a valid bonus percentage."

    def display_employee_details(self):
        return f"Employee ID: {self.employee_id}\nName: {self.name}\nSalary: {self.salary}"

if __name__ == "__main__":
    employee1 = Employee(101, "harsh tyagi", 50000)

    print(employee1.display_employee_details())  # Display employee details

    bonus_percentage = 10
    print(employee1.calculate_yearly_bonus(bonus_percentage))  # Calculate and display yearly bonus

    bonus_percentage = -5  # Attempt to use a negative bonus percentage
    print(employee1.calculate_yearly_bonus(bonus_percentage))
```

```python
# Problem 3: Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles.
```

```python
class VehicleRentalSystem:
    def __init__(self):
        self.available_vehicles = {}

    def add_vehicle(self, vehicle_id, vehicle_type):
        if vehicle_id not in self.available_vehicles:
            self.available_vehicles[vehicle_id] = vehicle_type
            return f"Vehicle {vehicle_type} (ID: {vehicle_id}) added to the rental system."
        else:
            return f"Vehicle {vehicle_type} (ID: {vehicle_id}) already exists in the rental system."

    def rent_vehicle(self, vehicle_id):
        if vehicle_id in self.available_vehicles:
            rented_vehicle = self.available_vehicles.pop(vehicle_id)
            return f"Vehicle (ID: {vehicle_id}) has been rented. Enjoy your {rented_vehicle}."
        else:
            return f"Vehicle (ID: {vehicle_id}) is not available for rent."

    def return_vehicle(self, vehicle_id, vehicle_type):
        if vehicle_id not in self.available_vehicles:
            self.available_vehicles[vehicle_id] = vehicle_type
            return f"Vehicle (ID: {vehicle_id}) has been returned to the rental system."
        else:
            return f"Vehicle (ID: {vehicle_id}) cannot be returned as it is not currently rented."

    def display_available_vehicles(self):
        if self.available_vehicles:
            available = ", ".join([f"{vehicle_id}: {vehicle_type}" for vehicle_id, vehicle_type in self.available_vehicles.items()])
            return f"Available Vehicles: {available}"
        else:
            return "No vehicles are available for rent at the moment."

if __name__ == "__main__":
    rental_system = VehicleRentalSystem()

    print(rental_system.add_vehicle("V001", "Sedan"))
    print(rental_system.add_vehicle("V002", "SUV"))

    print(rental_system.display_available_vehicles())

    print(rental_system.rent_vehicle("V001"))
    print(rental_system.display_available_vehicles())

    print(rental_system.return_vehicle("V001", "Sedan"))
    print(rental_system.display_available_vehicles())
```

```python
# Problem 4: Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.
```

```python
class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_available = True

    def __str__(self):
        return f"Title: {self.title}\nAuthor: {self.author}\nISBN: {self.isbn}\nAvailable: {'Yes' if self.is_available else 'No'}"


class Library:
    def __init__(self):
        self.catalog = {}

    def add_book(self, title, author, isbn):
        if isbn not in self.catalog:
            book = Book(title, author, isbn)
            self.catalog[isbn] = book
            return f"Book '{title}' by {author} (ISBN: {isbn}) added to the library catalog."
        else:
            return f"Book with ISBN {isbn} already exists in the catalog."

    def borrow_book(self, isbn):
        if isbn in self.catalog:
            book = self.catalog[isbn]
            if book.is_available:
                book.is_available = False
                return f"You have borrowed '{book.title}' by {book.author}."
            else:
                return f"The book '{book.title}' by {book.author} is currently not available."
        else:
            return f"Book with ISBN {isbn} is not in the catalog."

    def return_book(self, isbn):
        if isbn in self.catalog:
            book = self.catalog[isbn]
            if not book.is_available:
                book.is_available = True
                return f"Thank you for returning '{book.title}' by {book.author}."
            else:
                return f"The book '{book.title}' by {book.author} was not borrowed from this library."
        else:
            return f"Book with ISBN {isbn} is not in the catalog."

    def display_available_books(self):
        available_books = [book for book in self.catalog.values() if book.is_available]
        if available_books:
            return "\n".join([str(book) for book in available_books])
        else:
            return "No books are currently available in the library."

if __name__ == "__main__":
    library = Library()

    print(library.add_book("The Rich Dad Poor Dad", "F. Scott Fitzgerald", "978-0743273565"))
    print(library.add_book("To Kill a Mockingbird", "Harper Lee", "978-0061120084"))

    print(library.display_available_books())

    print(library.borrow_book("978-0743273565"))
    print(library.display_available_books())

    print(library.return_book("978-0743273565"))
    print(library.display_available_books())

    print(library.return_book("978-0743273565"))  # Trying to return an already returned book
```

```python
# Problem 5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.
```

```python
class Product:
    def __init__(self, product_id, name, price, quantity):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.quantity = quantity

    def __str__(self):
        return f"Product ID: {self.product_id}\nName: {self.name}\nPrice: {self.price:.2f}\nQuantity: {self.quantity}"


class InventorySystem:
    def __init__(self):
        self.inventory = {}

    def add_product(self, product_id, name, price, quantity):
        if product_id not in self.inventory:
            product = Product(product_id, name, price, quantity)
            self.inventory[product_id] = product
            return f"Product '{name}' (ID: {product_id}) added to the inventory."
        else:
            return f"Product with ID {product_id} already exists in the inventory."

    def update_product_quantity(self, product_id, quantity_change):
        if product_id in self.inventory:
            product = self.inventory[product_id]
            if product.quantity + quantity_change >= 0:
                product.quantity += quantity_change
                return f"Quantity of '{product.name}' (ID: {product_id}) updated to {product.quantity}."
            else:
                return f"Quantity change would result in a negative quantity, not allowed."
        else:
            return f"Product with ID {product_id} is not in the inventory."

    def display_available_products(self):
        available_products = [product for product in self.inventory.values() if product.quantity > 0]
        if available_products:
            return "\n".join([str(product) for product in available_products])
        else:
            return "No products are currently available in the inventory."

if __name__ == "__main__":
    inventory_system = InventorySystem()

    print(inventory_system.add_product(101, "Laptop", 799.99, 10))
    print(inventory_system.add_product(102, "Smartphone", 499, 20))

    print(inventory_system.display_available_products())

    print(inventory_system.update_product_quantity(101, -5))
    print(inventory_system.display_available_products())

    print(inventory_system.update_product_quantity(101, 15))  # Trying to make the quantity negative
    print(inventory_system.display_available_products())

    print(inventory_system.add_product(101, "Tablet", 299.99, 5))  # Adding a product with an existing ID
```
